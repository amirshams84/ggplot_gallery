```r
#####################
# Library Loading
#####################
library(tidyverse)
library(ggplot2)
library(cowplot)
library(gplots)
library(ComplexHeatmap)
library(ggcorrplot)
```

```r
#####################
# Config Theme
#####################
theme_set(theme_grey())
options(repr.plot.width=10, repr.plot.height=10)
```

```r
#####################
# Color Pallett
#####################
Color_Spectrum <- colorpanel(20, "blue","green","red")
GroupID_Color_List = c("Naive"="#88008A", "CCR6neg"="#FFA500", "CCR6low"="#0456f8", "CCR6int"="#1D8F06", "CCR6high"="#e51134")
GroupID_Order_List = c("Naive", "CCR6neg", "CCR6low", "CCR6int", "CCR6high")
GroupID_Label_List = c("Naive"="Naive", "CCR6neg"=expression(paste("CCR6"^neg)), "CCR6low"=expression(paste("CCR6"^low)), "CCR6int"=expression(paste("CCR6"^int)), "CCR6high"=expression(paste("CCR6"^high)))
```

```r
#####################
# Input/Output
#####################
expression_file_Path = "../Data/Fluidigm_Donor123_Log2_Expression.txt"
gene_list_file_Path = "../Data/Fluidigm_Donor123_Gene_List.txt"
metadata_file_Path = "../Data/Fluidigm_Donor123_sample_design.txt"
```

```r
#####################
# Read I/O
#####################
#READ I/O
expression_DF = read_tsv(expression_file_Path)
#head(expression_DF)
#dim(expression_DF)
gene_list_DF = read_tsv(gene_list_file_Path)
#head(gene_list_DF)
#dim(gene_list_DF)
metadata_DF = read_tsv(metadata_file_Path)
#head(metadata_DF)
#dim(metadata_DF)
```

```r
####################
#Data Wrangling
####################

#1-leftjoin with sample metadata
filtered_expression_DF = gene_list_DF %>% inner_join(expression_DF, by = c("GeneID" = "ID"))
#head(filtered_expression_DF)
#dim(filtered_expression_DF)

#2- Transpose
transposed_filtered_expression_DF = filtered_expression_DF %>%  gather(var, value, -GeneID) %>% spread(GeneID, value)
#head(transposed_filtered_expression_DF)
#dim(transposed_filtered_expression_DF)

metadata_transposed_filtered_expression_DF = metadata_DF %>% inner_join(transposed_filtered_expression_DF, by = c("SampleID" = "var"))
#head(metadata_transposed_filtered_expression_DF)

average_metadata_transposed_filtered_expression_DF = metadata_transposed_filtered_expression_DF %>% select(-c("SampleID")) %>% group_by(GroupID) %>% summarise_all(list(mean))
#head(average_metadata_transposed_filtered_expression_DF)


Gene_Correlation_Matrix <- round(cor(average_metadata_transposed_filtered_expression_DF[,-c(1:1)]), 2)
head(Gene_Correlation_Matrix)
Gene_Correlation_pValue_Matrix = cor_pmat(average_metadata_transposed_filtered_expression_DF[,-c(1:1)], method="pearson")
head(Gene_Correlation_pValue_Matrix)
#od = hclust(dist(Gene_Correlation_Dist))$order
#Gene_Correlation_Matrix = Gene_Correlation_Dist[od, od]
#head(Gene_Correlation_Matrix)
#Scaled_Gene_Correlation_Matrix = t(apply(Gene_Correlation_Matrix, 1, scale))
#head(Scaled_Gene_Correlation_Matrix)
```

```r
ggcorrplot(Gene_Correlation_Matrix, hc.order = TRUE, outline.col = "white", p.mat = Gene_Correlation_pValue_Matrix)
```

```r
Color_Spectrum <- colorpanel(20, "blue","white","red")
####################
# Heatmap
####################
correlation_heatmap_Object = Heatmap(Gene_Correlation_Matrix, 
        show_row_names = TRUE, 
        show_column_names = TRUE, 
        row_dend_reorder = FALSE, 
        column_dend_reorder = FALSE, 
        cluster_columns = TRUE,
        cluster_rows = TRUE,
        show_column_dend = TRUE,
        show_row_dend = TRUE,
        col = Color_Spectrum,
        #row_dend_width = unit(1.0, "cm"),
        #col_dend_width = unit(1.0, "cm"),
        clustering_distance_rows = "pearson",
        clustering_distance_columns = "pearson",
        clustering_method_rows = "complete",
        clustering_method_columns = "complete",
        rect_gp = gpar(col = "black", lwd = 0.5),
        heatmap_legend_param = list(col_fun = colors, title="", legend_height = unit(6, "cm"))
       )
correlation_heatmap_Object = plot_grid(grid.grabExpr(draw(correlation_heatmap_Object)), nrow=1)
```

```r
options(repr.plot.width=10, repr.plot.height=10, repr.plot.res=200)
correlation_heatmap_Object
```

```r
##################
# SAVE PDF
##################
ggsave(file="../Result/Fluidigm_scRNA_GeneCorrelation_Heatmap.png", device="png", plot=correlation_heatmap_Object, width = 10, height = 10, units = "in", dpi=600)
```
