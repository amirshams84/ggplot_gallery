## Boxplot + kruskal_wallis + order

```r
#####################
# Library Loading
#####################
library(tidyverse)
library(ggpubr)
library(ggsci)
library(gtools)
```

```r
PLOTLY_Pallett = c('#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2','#7f7f7f','#bcbd22','#17becf')
JCO_PALLETT = pal_jco("default")(10)
FEATURE_PALLETT = PLOTLY_Pallett
```

```r
#####################
# Input/Output
#####################
group = "bacterial_species"
metadata_file_Path = "../Data/Processed_metadata/cross_sectional_plus_processed.txt"
abundance_file_Path = paste("../Data/Processed_relative_abundance/core_biom.", group, ".relative_abundance.txt", sep="")
boxplot_output = paste("../Result/core_biom.", group, ".boxplot.pdf", sep="")
```

```r
#####################
# Read I/O
#####################
abundance_DF = read_tsv(abundance_file_Path)
colnames(abundance_DF)[colnames(abundance_DF)=="#OTU ID"] <- "OTU_ID"
#head(abundance_DF)
metadata_DF = read_tsv(metadata_file_Path)
#colnames(metadata_DF)[colnames(metadata_DF)=="Design"] <- "State"
head(metadata_DF)
```

```r
metadata_DF %>% group_by(Grouping_Cross_Sectional) %>% tally()
```

```r
#################
#PROCESSED Abundance
#################
processed_abundance_DF = abundance_DF %>%
    gather(Sample_ID, Expression, -OTU_ID)
    
head(processed_abundance_DF)
```

```r
metadata_processed_abundance_DF = metadata_DF %>%
    select(Sample_ID, Grouping_Cross_Sectional) %>%
    inner_join(processed_abundance_DF, by = c("Sample_ID", "Sample_ID")) %>%
    mutate(log2_expression = log2(Expression +1))
head(metadata_processed_abundance_DF)
```

```r
total_abundance = metadata_processed_abundance_DF %>%
    group_by(OTU_ID) %>%
    summarize(total = sum(log2_expression))
head(total_abundance)
```

```r
total_metadata_processed_abundance_DF = metadata_processed_abundance_DF %>%
    inner_join(total_abundance, by = c("OTU_ID", "OTU_ID"))
head(total_metadata_processed_abundance_DF)
```

```r
top25_OTU = metadata_processed_abundance_DF %>%
    group_by(OTU_ID) %>%
    summarize(total = sum(log2_expression)) %>%
    mutate(median_total = median(total), Quantile = quantile(total, probs = c(0.75))) %>%
    filter(total >= Quantile) %>%
    top_n(25, total) %>%
    arrange(desc(total))
head(top25_OTU)
count = top25_OTU %>% tally()
```

```r
selected_metadata_processed_abundance_DF = total_metadata_processed_abundance_DF %>%
    filter(OTU_ID %in% top25_OTU$OTU_ID)
head(selected_metadata_processed_abundance_DF)
```

```r
####################
#FEATURE ORDER/COLOR
####################
Feature_Order_List = rev(unique(metadata_processed_abundance_DF[["Grouping_Cross_Sectional"]]))
Feature_Order_List = c("Healthy_Control","Compensated_Cirrhosis", "Cirrhosis_Lact", "Cirrhosis_Lact_Rif")
Feature_List_Size = length(Feature_Order_List)
Feature_Label_Dict = Feature_Order_List
names(Feature_Label_Dict) = Feature_Order_List
Feature_Order_Factor = factor(Feature_Order_List, levels=Feature_Order_List)
if(Feature_List_Size > 10){
    Feature_Color_List = colorRampPalette(FEATURE_PALLETT)(Feature_List_Size)
}else{
    Feature_Color_List = FEATURE_PALLETT[1:Feature_List_Size]
}
Feature_Color_Dict = Feature_Color_List
names(Feature_Color_Dict) = Feature_Order_List
#Feature_Color_Dict
#Feature_Order_List
```

```r
options(repr.plot.width=10, repr.plot.height=5, repr.plot.res=200)
####################
# BoxPlot
####################
BoxPlot_Object = ggplot(
    selected_metadata_processed_abundance_DF, 
    aes(x=reorder(OTU_ID, -Expression), y=Expression, label=OTU_ID, fill =factor(Grouping_Cross_Sectional, levels = rev(Feature_Order_List)))) +
geom_boxplot(position = position_dodge2(reverse = TRUE), size = 0.15, fatten = 0.5, outlier.alpha = 0.2, outlier.size = 0.5) +    
#geom_boxplot(stat = "boxplot", position = position_dodge2(reverse = TRUE), #aes(fill=Grouping_Cross_Sectional), outlier.shape = 21, outlier.alpha = 0.2, outlier.color = "black", size = 0.15, show.legend=TRUE, color="black", fatten = 0.5,) +
    #geom_point(position=position_jitterdodge(), aes(fill=Grouping_Cross_Sectional), shape=21, stroke = 0.2) + 
    #stat_compare_means(label = "p.signif", label.x = 1.5, label.y = 40) +


    stat_compare_means(method = "kruskal.test", label = "p.signif", hide.ns=TRUE,
                      symnum.args = list(cutpoints = c(0, 0.0001, 0.001, 0.01, 0.05, 1), 
                        symbols = c("****", "***", "**", "*", "ns"))
                      ) +
    #scale_x_discrete(breaks = Feature_Order_List, labels=Feature_Order_List, limits=Feature_Order_List) +
    scale_fill_manual(breaks = Feature_Order_List, labels=Feature_Label_Dict, limits=Feature_Order_List, values=Feature_Color_Dict) +
    labs(title = paste(group,"[top", count, "]",sep=""), subtitle = NULL, caption = NULL, tag = NULL, x=NULL, y="Abundance") +
    #scale_y_continuous(breaks=c(0.25, 0.5, 0.75, 1.00)) +
    scale_y_sqrt() +
#theme_minimal() +
    theme(
        #Title
    text=element_text(size=8,  family="sans", face="plain"),
    plot.title = element_text(size=12,  family="sans", face="plain", hjust = 0.0),
    plot.subtitle = element_text(size=10,  family="sans", face="plain", hjust = 0.5),
    plot.caption = element_text(size=8,  family="sans", face="plain", color = "black", hjust = 1),
    #Axis
    axis.title.x = element_text(size=12,  family="sans", face="plain", color = "black", hjust = 0.5),
    axis.title.y = element_text(size=10,  family="sans", face="plain", color = "black", hjust = 0.5, vjust=0),
    axis.text.x = element_text(size=10,  family="sans", face="plain", color = "black", angle=-45, hjust=0.0,vjust=0.5),
    axis.text.y = element_text(size=8,  family="sans", face="plain", color = "black", angle=0, vjust=0.5),
    #axis.line = element_line(colour = "black"),
    #Panel
    panel.grid.major = element_line(color = "grey95", size = 0.25), 
    panel.grid.minor = element_line(color = "grey95", size = 0.25),
    panel.background = element_rect(color = "black", size = 0.25, fill = "grey99"),
    panel.border = element_rect(color = "black", size = 0.25, fill = NA),
    #Strip
    strip.text = element_text(size=10,  family="sans", face="plain", color = "black"),
    strip.text.x = element_text(size=12,  family="sans", face="plain", color = "black"),
    strip.text.y = element_text(size=12,  family="sans", face="plain", color = "black"),
    strip.background = element_rect(color = "black", size = 0.25, fill = "grey95"),
    #Legend
    legend.position="right",
    legend.title = element_blank(),
    legend.key=element_blank(),
    legend.text = element_text(size=8,  family="sans", face="plain", color = "black", angle=0, vjust=0.5),
    legend.background = element_blank(),
    
    )
BoxPlot_Object
```

```r
ggsave(file=boxplot_output, device=cairo_pdf, plot=BoxPlot_Object, width = 12.5, height =5, units = "in", dpi=1200, limitsize = FALSE)
```
