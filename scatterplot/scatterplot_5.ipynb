```r
#####################
# Library Loading
#####################
library(scales)
library(ggpmisc)
library(tidyverse)
library(cowplot)
library(readxl)

```

```r
#####################
# Config Theme
#####################
theme_set(theme_minimal())
options(repr.plot.width=10, repr.plot.height=10, repr.plot.res=200)
```

```r
#####################
# Input/Output
#####################
regression_file_Path = "../Data/Fluidigm_Donor123_regression.xlsx"
```

```r
#####################
# Read I/O
#####################
regression_DF = read_excel(regression_file_Path)
head(regression_DF)
```

```r
#####################
# Data Wrangling
#####################
gene_List = list()
gene_List[[1]] = c("RORC vs IL17a","CCL20 vs IL17a","IL22 vs IL17a")
gene_List[[2]] = c("GATA3 vs IL17a","IL4 vs IL17a","IL13 vs IL17a")
gene_List[[3]] = c("IFNG vs IL17a","TBx21 vs IL17a","EOMES vs IL17a")

gene_label_List = list()
gene_label_List[[1]] = c("RORC vs IL17a"="RORC expression level", "CCL20 vs IL17a"="CCL20 expression level", "IL22 vs IL17a"="IL22 expression level")
gene_label_List[[2]] = c("GATA3 vs IL17a"="GATA3 expression level", "IL4 vs IL17a"="IL4 expression level", "IL13 vs IL17a"="IL13 expression level")
gene_label_List[[3]] = c("IFNG vs IL17a"="IFNG expression level", "TBx21 vs IL17a"="TBx21 expression level", "EOMES vs IL17a"="EOMES expression level")

gather_regression_DF = regression_DF %>%
    gather(variable, value,-Level) %>%
    filter(Level %in% c(1,2,3))
# ----------
my.formula <- y ~ x
# ----------




regressionPlot_Object_List = list()
for(each_index in 1:length(gene_List)){
    ##
    filter_gather_regression_DF = gather_regression_DF %>%
        filter(variable %in% gene_List[[each_index]])
    #print(head(filter_gather_regression_DF))
    
    if((each_index %% 2) == 0) {
        selected_Lab = labs(title = NULL, subtitle = NULL, caption = NULL, tag = NULL, x=NULL, y="Frequency of IL17 expressing cells")
    } else {
        selected_Lab = labs(title = NULL, subtitle = NULL, caption = NULL, tag = NULL, x=NULL, y="")
    }
    
    Gene_Order_List = list(unique(filter_gather_regression_DF$variable))
    Gene_order_factor = factor(Gene_Order_List, levels=Gene_Order_List)
    #
    Regression_Plot_Object = ggplot(filter_gather_regression_DF, aes(x=Level, y=value)) +
        geom_point(color="darkblue", size=2) +
        geom_smooth(method = "lm", se=FALSE,formula = my.formula, size=0.5) +
        stat_poly_eq(position = "identity",formula = my.formula, parse = TRUE,coef.digits = 3, rr.digits = 4, label.x = "right", label.y = "top", size = 5) +
        labs(title = NULL, subtitle = NULL, caption = NULL, tag = NULL, x=NULL, y="Frequency of IL17 expressing cells") +
        scale_x_continuous(breaks = scales::pretty_breaks(n = 5)) +
        scale_y_continuous(breaks = scales::pretty_breaks(n = 5)) +
        theme(
            #Title
            legend.position="none",
            panel.border = element_rect(color = "black", fill = NA),
            strip.text.x = element_text(size=14,  family="sans", face="plain", color = "black", hjust = 0.5),
            strip.text.y = element_text(size=14,  family="sans", face="plain", color = "black", hjust = 0.5),
            axis.text.x = element_text(size=8,  family="sans", face="plain", color = "black", angle=0, vjust=0.5),
            axis.text.y = element_text(size=8,  family="sans", face="plain", color = "black", angle=0, vjust=0.5),
            axis.title.y = element_text(size=14,  family="sans", face="plain", color = "black"),
            #plot.margin=margin(l=0.5,t=0,r=0,b=0,unit="cm"),
            aspect.ratio=1,
            strip.placement = "outside"
        ) +
        facet_wrap(~factor(variable, 
                   levels=gene_List[[each_index]], 
                   labels=gene_label_List[[each_index]]), nrow=1, scales="free", strip.position = "bottom")
    
    regressionPlot_Object_List[[each_index]] = plot_grid(Regression_Plot_Object, nrow=1)
}
```

```r
options(repr.plot.width=10, repr.plot.height=10, repr.plot.res=200)
Regression_Plot = plot_grid(plotlist=regressionPlot_Object_List, ncol = 1)
Regression_Plot
```

```r
##### # SAVE PDF
##################
ggsave(file="../Result/Fluidigm_scRNA_Regression.png", device="png", plot=Regression_Plot, width = 10, height = 10, units = "in", dpi=600)
```
