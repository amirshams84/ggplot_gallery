```r
#####################
# Data Loading
#####################
library(tidyverse)
library(ggplot2)
library(reshape2)
library(cowplot)
library(tools)
library(scales)
library(stringr)
######################
#SETTING
args = commandArgs(trailingOnly=TRUE)
options(scipen=10000000)
options(dplyr.width = Inf)
Default_Color = c('#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2','#7f7f7f','#bcbd22','#17becf')
theme_set(theme_bw(base_family = "Sans"))
OFFLINE_MODE=FALSE
if(OFFLINE_MODE==FALSE){
options(repr.plot.width=10, repr.plot.height=10)
}
######################
#INPUT/OUTPUT
if(OFFLINE_MODE==FALSE){
    abundance_file_Path = "/data/shamsaddinisha/MBAC/MOTHUR/MISEQ_SOP/MISEQ_SOP_SMALL/Late/cluster/Late.OTU_abundance.txt"
    info_file_Path = "/data/shamsaddinisha/MBAC/MOTHUR/MISEQ_SOP/MISEQ_SOP_SMALL/Late/overview/Late.overview.duplicate_count.plot.txt"
    metadata_file_Path = "/data/shamsaddinisha/MBAC/MOTHUR/MISEQ_SOP/MISEQ_SOP_SMALL/Late/overview/Late_mothur_metadata.txt"
    mapped_count_plot_output = "./Late.overview.mapped_count.txt"
}else{
    abundance_file_Path = args[1]
    info_file_Path = args[2]
    metadata_file_Path = args[3]
    mapped_count_plot_output = args[4]
}
######################
#READ I/O
abundance_DF = read_tsv(abundance_file_Path)
#head(abundance_DF)
#dim(abundance_DF)
metadata_DF = read_tsv(metadata_file_Path)
#head(metadata_DF)
#dim(metadata_DF)
info_file_DF = read_tsv(info_file_Path)
#head(info_file_DF)
#dim(metadata_DF)
####################
#Data Wrangling
colnames(abundance_DF)[colnames(abundance_DF)=="#OTU ID"] <- "OTU_ID"
#
design_List = sort(unique(metadata_DF$Treatment))
design_list_Length = length(design_List)
#
melted_abundance_DF = melt(abundance_DF, id.vars = c("OTU_ID"), variable.name = "Sample_ID", value.name="Abundance")
metadata_melted_abundance_DF = metadata_DF %>%
    left_join(melted_abundance_DF, by = c("Sample_ID" = "Sample_ID")) %>%
    filter(Treatment == design_List[[1]]) %>%
    group_by(Treatment, Sample_ID) %>%
    summarise(mapped_count = sum(Abundance)) %>%
    arrange(-mapped_count)
#
info_metadata_melted_abundance_DF = info_file_DF %>% select(-c(Treatment)) %>%
    left_join(metadata_melted_abundance_DF, by = c("Sample_ID" = "Sample_ID")) %>% 
    mutate(unmapped_count = read_count-mapped_count) %>%
    select(-c(unique_count, read_count, duplicate_count))
#
sample_count = length(unique(info_metadata_melted_abundance_DF$Sample_ID))
if(sample_count <= 20){
    sample_count_limit = sample_count 
}else{
    sample_count_limit = 20
}
filtered_info_metadata_melted_abundance_DF = info_metadata_melted_abundance_DF %>% arrange(desc(mapped_count)) %>% top_n(sample_count, mapped_count)
#
melted_info_metadata_melted_abundance_DF = melt(filtered_info_metadata_melted_abundance_DF, id.vars = c("Sample_ID", "Treatment"), variable.name = "Variable", value.name="Value")
#
Color_List = c("mapped_count"='#1f77b4', "unmapped_count"='#ff7f0e')
Order_List = c("mapped_count", "unmapped_count")
Label_List = c("mapped_count"="Mapped Reads", "unmapped_count"="Unmapped Reads")
factor_order = factor(Order_List, levels=Order_List)
#
melted_info_metadata_melted_abundance_DF
#####################
#Data Plotting
ggplot_Object = ggplot(melted_info_metadata_melted_abundance_DF, aes(x = reorder(Sample_ID, -Value), y = Value,  fill = forcats::fct_rev(Variable))) + 
    geom_bar(stat="identity",color="black", size=0.25) +
    geom_text(aes(label = scales::comma(Value)),position = position_stack(vjust = .5), size=4) +
    scale_color_manual(values = colorRampPalette(Default_Color)(sample_count)) +
    scale_y_continuous(labels = scales::comma, breaks = scales::pretty_breaks(n = 20)) +
    scale_fill_manual(breaks = Order_List, labels=Label_List, limits=Order_List, values=Color_List) +
    labs(title = design_List[[1]], caption = NULL, subtitle = NULL, tag = NULL, x=NULL, y=NULL) + 
    theme(
        text=element_text(family="Sans"),
        plot.title = element_text(size=14, color = "black", face = "plain" ),
        plot.subtitle = element_text(size=12, color = "black", face = "plain" ),
        plot.caption = element_text(size=10, color = "black", face = "italic"),
        #Axis
        axis.title.x = element_text(size=10,  family="sans", face="plain", color = "black", hjust = 0.5),
        axis.title.y = element_text(size=10,  family="sans", face="plain", color = "black", hjust = 0.5, vjust=0),
        axis.text.x = element_text(size=10,  family="sans", face="plain", color = "black", angle=45, vjust=0.5),
        axis.text.y = element_text(size=8,  family="sans", face="plain", color = "black", angle=0, vjust=0.5),
        #LEGEND
        #legend.position="none",
        legend.position="right",
        legend.title = element_blank(),
        legend.key=element_blank(),
        legend.text.align = 0,
        legend.text = element_text(size=10,  family="sans", face="plain", color = "black", angle=0, vjust=0.5),
        legend.background = element_blank(),
    )
if(OFFLINE_MODE==FALSE){
ggplot_Object
}
if(OFFLINE_MODE==TRUE){

#####################
#EXPORTING
file_extension = file_ext(mapped_count_plot_output)
#
plot_PDF = gsub(file_extension, "plot.pdf", mapped_count_plot_output)
plot_PNG = gsub(".pdf", ".png", plot_PDF)
plot_JPG = gsub(".pdf", ".jpg", plot_PDF)
plot_SVG = gsub(".pdf", ".svg", plot_PDF)
plot_TXT = gsub(".pdf", ".txt", plot_PDF)
#
ggsave(file=plot_PDF, device=cairo_pdf, plot=ggplot_Object, width=15, height =5, units = "in", limitsize = FALSE, dpi=1200)
ggsave(file=plot_PNG, device="png", plot=ggplot_Object, width=15, height =5, units = "in", limitsize = FALSE, dpi=300)
ggsave(file=plot_JPG, device="jpg", plot=ggplot_Object, width=15, height =5, units = "in", limitsize = FALSE, dpi=300)
ggsave(file=plot_SVG, device="svg", plot=ggplot_Object, width=15, height =5, units = "in", limitsize = FALSE, dpi=72)
write_tsv(path=plot_TXT, x=info_metadata_melted_abundance_DF)
print("Mapped count plot completed Successfully!!!")
}
```
