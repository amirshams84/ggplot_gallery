```r
#####################
# Library Loading
#####################
library(tidyverse)
library(cowplot)
library(readxl)
library(plot3D)
```

```r
#####################
# Config Theme
#####################
theme_set(theme_minimal())
```

```r
#####################
# Color Pallett
#####################
GroupID_Color_List = c("Naive"="#88008A", "CCR6neg"="#FFA500", "CCR6low"="#0456f8", "CCR6int"="#1D8F06", "CCR6high"="#e51134")
GroupID_Order_List = c("Naive", "CCR6neg", "CCR6low", "CCR6int", "CCR6high")
#GroupID_Order_List = c("Naive", "CCR6neg", "CCR6low", "CCR6int", "CCR6high")
GroupID_Label_List = c("Naive"="Naive", "CCR6neg"=expression(paste("CCR6"^neg)), "CCR6low"=expression(paste("CCR6"^low)), "CCR6int"=expression(paste("CCR6"^int)), "CCR6high"=expression(paste("CCR6"^high)))
```

```r
#####################
# Input/Output
#####################
PCA_file_Path = "../Data/Fluidigm_Donor123_PCA.xlsx"
gene_list_file_Path = "../Data/Fluidigm_Donor123_Gene_List.txt"
metadata_file_Path = "../Data/Fluidigm_Donor123_sample_design.txt"
```

```r
#####################
# Read I/O
#####################
PCA_sample_score_DF = read_excel(PCA_file_Path, sheet= "sample_scores")
#head(PCA_sample_score_DF)
#dim(PCA_sample_score_DF)
gene_list_DF = read_tsv(gene_list_file_Path)
#head(gene_list_DF)
#dim(gene_list_DF)
metadata_DF = read_tsv(metadata_file_Path)
#head(metadata_DF)
#dim(metadata_DF)
```

```r
####################
#PCA
####################
#head(PCA_sample_score_DF)
PCA_var = PCA_sample_score_DF %>% select(-c("SampleID", GroupID)) %>% summarise_all(list(var))
PCA_sum = sum(PCA_var)
######
```

```r
processed_PCA_sample_score_DF = PCA_sample_score_DF %>% mutate(Tretment_Level = as.numeric(factor(GroupID)))
#head(processed_PCA_sample_score_DF)
hash_Group = sapply(processed_PCA_sample_score_DF$GroupID, match, unique(processed_PCA_sample_score_DF$GroupID))
color_List = c("#88008A", "#FFA500", "#0456f8", "#1D8F06", "#e51134")
#head(unique(hash_Group))
color_order = unique(processed_PCA_sample_score_DF$GroupID)
#color_order
#unique(hash_Group)
legend_name_placement = unique(hash_Group)
legend_name_placement = c(4.5,4,3.5,3,2.5)
#pdf(file="../Result/Fluidigm_Dono123_PCA_3D.pdf", height = 10, width = 10, useDingbats=FALSE, pointsize=1/300)
```

```r
PC1_percentage = round((PCA_var$PC1 / PCA_sum) * 100, 2)
PC1_title = paste("PC1 (",as.character(PC1_percentage), "%)", sep="")
PC2_percentage = round((PCA_var$PC2 / PCA_sum) * 100, 2)
PC2_title = paste( "PC2 (",as.character(PC2_percentage),"%)", sep="")
PC3_percentage = round((PCA_var$PC3 / PCA_sum) * 100, 2)
PC3_title = paste( "PC3 (",as.character(PC3_percentage),"%)", sep="")
```

```r
options(repr.plot.width=10, repr.plot.height=10, repr.plot.res=200)
options(scipen=10000)

PCA_scatter_3D = scatter3D(
    x=processed_PCA_sample_score_DF$PC1, 
    y=processed_PCA_sample_score_DF$PC2, 
    z=processed_PCA_sample_score_DF$PC3, 
    col=rev(color_List[1:length(unique(hash_Group))]),
    colvar = hash_Group,
    clab = NA,
    colkey = FALSE,
    #colkey =list(at = c(4.5,3.8,3,2.2,1.5),
    #             side = 4, 
    #             length = 0.2,
    #             width = 0.5, 
    #             #dist = -0.03,
    #             #font.lab = 20,#legend title font
    #            addlines = TRUE,
    #             tick=FALSE, 
    #             cex.axis = 0.8, 
    #             cex.clab = 3,
    #             #col.lab=2,
    #             tick = FALSE,
    #             labels = GroupID_Label_List),
    #
    xlab = PC1_title, 
    ylab = PC2_title, 
    zlab = PC3_title,
    #
    ############
    #tick
    #########
    xaxt="n", 
    yaxt="n",
    ticktype = "simple",
    #ticktype = "detailed",
    #tick = NA,
    nticks=4,
    tck=-0.01,
    cex.axis = 1, #font of axis tick
    cex.lab = 2, #font of axis text
    bty = "g",
    #bty = "f",
    col.axis = "black",
   
    
    #cex.font=10,
    
    #col.panel = "lightgrey",
    
    #font.main = "arial",
    col.grid = "white", 
    lwd.panel=1,
    lwd.grid = 0.1,
    pch = 20,#pointshape
    cex=2,#point_size
    type = "p",
    phi = -90,
    theta =0,
    main=NULL,
    plot=TRUE,
    rasterImage = TRUE
)
#dev.off()
```
